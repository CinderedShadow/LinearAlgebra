```python
def transpose(matrix):
  rows=len(matrix)
  cols=len(matrix[0])
  tempMatrix=[]
  for col in range(cols):
      colList=[]
      for row in range(rows):
          colList.append(matrix[row][col])
      tempMatrix.append(colList)
  return tempMatrix

def getRowExchangeMatrix(matrix,exchangeRows): #exhangeRows is a tuple
    swapRow1,swapRow2=exchangeRows
    swapRow1,swapRow2=swapRow1-1,swapRow2-1
    rows=len(matrix)
    tempMatrix=[[0 for _ in range(rows)] for _ in range(rows)]
    rows=len(tempMatrix)
    cols=len(tempMatrix[0])
    for row in range(rows):
        for col in range(cols):
            if row==col and row!=swapRow1 and row!=swapRow2:
                tempMatrix[row][col]=1
            if row==swapRow1 and col==swapRow2:
                tempMatrix[row][col]=1
            if row==swapRow2 and col==swapRow1:
                tempMatrix[row][col]=1
    return tempMatrix

def rowExchange(matrix,exchangeRows):
    rowExchangeMatrix=getRowExchangeMatrix(matrix,exchangeRows)
    resultMatrix=matrixMultiplication(rowExchangeMatrix,matrix)
    return resultMatrix


matrix1 = [
    [1, 2],
    [3, 4]
]

matrix2 = [
    [5],
    [7]
]

t=(1,2)
rowExchange(matrix2,t)
```

```python
def getRowReplacementMatrix(matrix,replacementRows,scale): #replacementRows is a tuple
    addRow1,addRow2=replacementRows
    addRow1,addRow2=addRow1-1,addRow2-1
    rows=len(matrix)
    tempMatrix=[[0 for _ in range(rows)] for _ in range(rows)]
    rows=len(tempMatrix)
    cols=len(tempMatrix[0])
    for row in range(rows):
        for col in range(cols):
            if row==col:
                tempMatrix[row][col]=1
            if row==addRow2 and col==addRow1:
                tempMatrix[row][col]=scale
    return tempMatrix

def rowReplacement(matrix,replacementRows,scale=1):
    rowReplacementMatrix=getRowReplacementMatrix(matrix,replacementRows,scale)
    resultMatrix=matrixMultiplication(rowReplacementMatrix,matrix)
    return resultMatrix


matrix1 = [
    [1, 2],
    [3, 4]
]

matrix2 = [
    [5],
    [7]
]

t=(1,2)
rowReplacement(matrix1,t,5)
```

```python
def getRowScalingMatrix(matrix,scale_Row): #scaleRows is a tuple= (scale, row)
    scale,rowScale=scale_Row
    rowScale-=1
    rows=len(matrix)
    tempMatrix=[[0 for _ in range(rows)] for _ in range(rows)]
    rows=len(tempMatrix)
    cols=len(tempMatrix[0])
    for row in range(rows):
        for col in range(cols):
            if row==col and row!=rowScale:
                tempMatrix[row][col]=1
            if row==col and row==rowScale:
                tempMatrix[row][col]=scale
    return tempMatrix

def rowScaling(matrix,scale_Row):
    rowScalingMatrix=getRowScalingMatrix(matrix,scale_Row)
    resultMatrix=matrixMultiplication(rowScalingMatrix,matrix)
    return resultMatrix


matrix1 = [
    [1, 2],
    [3, 4]
]

matrix2 = [
    [5],
    [7]
]

t=(3,1)
rowScaling(matrix2,t)
```

```python
import copy

def isEchelonForm(matrix):
  rows = len(matrix)
  cols = len(matrix[0])
  prevLeadingIndex = -1

  for row in range(rows):
      rowList = matrix[row]
      leadingIndex = None

      for col in range(cols):
          if matrix[row][col] != 0:
              leadingIndex = col
              break

      if leadingIndex is None:
          if prevLeadingIndex != -1:
              continue
      else:
          if leadingIndex <= prevLeadingIndex:
              return False
          prevLeadingIndex = leadingIndex

  for i in range(1, rows):
      if all(x == 0 for x in matrix[i-1]) and any(matrix[i]):
          return False
  return True

def getEchelonForm(originalMatrix): #Returns EF and elementary matrix
    matrix = copy.deepcopy(originalMatrix)
    rows = len(matrix)
    cols = len(matrix[0])
    #Make elementary transformation matrix
    ElementaryMatrix=[[0 for v in range(rows)] for i in range(rows)]
    for row in range(rows):
        for col in range(rows):
            if row==col:
                ElementaryMatrix[row][col]=1
    currentPivot = 0
    
    for r in range(rows):
        pivot_found = False
        for i in range(r, rows):
            if matrix[i][currentPivot] != 0:
                matrix=rowExchange(matrix,(i,r))
                ElementaryMatrix=matrixMultiplication(getRowExchangeMatrix(matrix,(i,r)),ElementaryMatrix)
                pivot_found = True
                break
        
        if not pivot_found:
            currentPivot += 1
            if currentPivot >= cols:
                break
            continue

        for i in range(r + 1, rows):
            if matrix[i][currentPivot] != 0:
                factor = matrix[i][currentPivot] / matrix[r][currentPivot]
                matrix = rowReplacement(matrix,(r+1,i+1),-factor)
                ElementaryMatrix=matrixMultiplication(getRowReplacementMatrix(matrix,(r+1,i+1),-factor),ElementaryMatrix)
        
        currentPivot += 1
        if currentPivot >= cols:
            break
    return prettyMatrix(matrix),ElementaryMatrix

def prettyMatrix(matrix):
    rows=len(matrix)
    cols=len(matrix[0])
    for row in range(rows):
        for col in range(cols):
            if matrix[row][col]==int(matrix[row][col]):
                matrix[row][col]=int(matrix[row][col])
            else:
                matrix[row][col] = round(matrix[row][col], 1)
    return matrix


def getReducedEchelonForm(originalMatrix): #Returns REF and elementary matrix
    matrix = copy.deepcopy(originalMatrix)
    echelonForm,ElementaryMatrix = getEchelonForm(matrix)
    rows = len(echelonForm)
    cols = len(echelonForm[0])
    
    for rowIndex in range(rows - 1, -1, -1):  # go through rows from bottom to top
        pivotCol = None
        for colIndex in range(cols):
            if echelonForm[rowIndex][colIndex] != 0:
                pivotCol = colIndex
                break
        
        if pivotCol is None:
            continue

        pivot = echelonForm[rowIndex][pivotCol]
        echelonForm = rowScaling(echelonForm,(1/pivot,rowIndex+1))
        ElementaryMatrix=matrixMultiplication(getRowScalingMatrix(echelonForm,(1/pivot,rowIndex+1)),ElementaryMatrix)

        for upperRowIndex in range(rowIndex - 1, -1, -1):
            factor = echelonForm[upperRowIndex][pivotCol]
            echelonForm=rowReplacement(echelonForm,(rowIndex+1,upperRowIndex+1),-factor)
            ElementaryMatrix=matrixMultiplication(getRowReplacementMatrix(echelonForm,(rowIndex+1,upperRowIndex+1),-factor),ElementaryMatrix)


    for rowIndex in range(rows):
        rowList=echelonForm[rowIndex]
        if isEmptyRow(rowList):
            nextRow=rowIndex+2
            if nextRow>rows:
                nextRows=rows
            echelonForm=echelonForm=rowExchange(echelonForm,(rowIndex+1,rowIndex+2))
            ElementaryMatrix=matrixMultiplication(getRowExchangeMatrix(echelonForm,(rowIndex+1,rowIndex+2)),ElementaryMatrix)
    
    return prettyMatrix(prettyMatrix(echelonForm)),ElementaryMatrix

def isEmptyRow(row):
    for element in row:
        if element != 0:
            return False
    return True

A = [
    [1, 2, 2, 2, 2],
    [2, 4, 6, 8, 10],
    [3, 6, 8, 10, 12],
    [-1, -2, -2, -1, -2]
]

print("Here is the Echelon Form!")
EF,E=getEchelonForm(A)
prettyPrint(EF)
print("Here is the Reduced Echelon Form!")
REF,E=getReducedEchelonForm(A)
prettyPrint(REF)

```

```python
def getNullSpace(matrix):
    ref, e = getReducedEchelonForm(matrix)
    rows, cols = len(ref), len(ref[0])  
    
    pivotCols = []
    for i in range(rows):
        for j in range(cols):
            if ref[i][j] != 0:
                pivotCols.append(j)
                break
    
    freeVars = [j for j in range(cols) if j not in pivotCols]
    
    nullSpaceBasis = []
    for freeVar in freeVars:
        basisVector = [0] * cols
        basisVector[freeVar] = 1 
        
        for i in range(len(pivotCols) - 1, -1, -1):
            pivotCol = pivotCols[i]
            basisVector[pivotCol] = -sum(
                ref[i][j] * basisVector[j] for j in range(pivotCol + 1, cols)
            ) / ref[i][pivotCol]
        
        nullSpaceBasis.append(basisVector)

    if nullSpaceBasis == []:
        nullSpaceBasis = [[0 for _ in range(cols)]]

    return prettyMatrix(nullSpaceBasis)

def getColumnSpace(matrix):
    ref, e = getReducedEchelonForm(matrix)
    rows, cols = len(ref), len(ref[0])

    pivotCols = []
    for i in range(rows):
        for j in range(cols):
            if ref[i][j] != 0:
                pivotCols.append(j)
                break

    if not pivotCols:
        columnSpaceBasis = [[0] * rows]
    else:
        columnSpaceBasis = [[matrix[i][col] for i in range(len(matrix))] for col in pivotCols]
    return prettyMatrix(columnSpaceBasis)


A = [
    [1, 0, 0],
    [0, 1, 0],
    [0, 0, 1],
    [0, 0, 0]
]

A = [
    [1, 2, 2, 2, 2],
    [2, 4, 6, 8, 10],
    [3, 6, 8, 10, 12],
    [-1, -2, -2, -1, -2]
]

nullSpace = getNullSpace(A)
print("Basis for Null Space:")
print(nullSpace)

columnSpace = getColumnSpace(A)
print("Basis for Column Space:")
print(columnSpace)


def getMatrixSolutions(A, b):
    refA, E = getReducedEchelonForm(A)
    refB = prettyMatrix(matrixMultiplication(E, b))
    particularSolution=[]
    
    for row in range(len(refA)):
        rowList=refA[row]
        if isEmptyRow(rowList) and refB[row] != [0]:
            return None 


    columnSpaceBasis= getColumnSpace(A)
    nullSpaceBasis = getNullSpace(A)
    AT=transpose(A)
    prettyPrint(AT)
    cols=len(AT)
    for col in range(cols):
        colList=AT[col]
        if colList in columnSpaceBasis:
            particularSolution.append(refB[0])
            refB.pop(0)
        else:
            particularSolution.append([0])


    finalParticularSolution=[]
    for i in particularSolution:
        finalParticularSolution.append(i[0])

    
    nullSpaceStr = f"{{ {', '.join(str(v) for v in nullSpaceBasis)} }}"
    
    return f"Solution: {finalParticularSolution} + span({nullSpaceStr})"



A = [
    [1, 2, 2, 2, 2],
    [2, 4, 6, 8, 10],
    [3, 6, 8, 10, 12],
    [-1, -2, -2, -1, -2]
]


b=[[1],
   [5],
   [6],
   [1]]

print(getMatrixSolutions(A,b))
```

```python
import inspect
import sys
import random
import types

from io import StringIO

EPSILON = 10**-7


def almostEqual(x, y, epsilon=EPSILON):
    return abs(x - y) <= epsilon


def rounded(d):
    sign = 1 if (d >= 0) else -1
    d = abs(d)
    n = int(d)
    if (d - n >= 0.5):
        n += 1
    return sign * n


def testFunction(fn):
    return fn


# Capturing print output
tempOut = StringIO()


def startOutputCapture():
    global tempOut
    tempOut = StringIO()
    sys.stdout = tempOut


def stopOutputCapture():
    output = tempOut.getvalue()
    sys.stdout = sys.__stdout__
    return output


# Faking input
fakeInputs = []


def fakeInput(prompt=''):
    if fakeInputs == []:
        raise Exception('input() was called, but there are no more inputs left.')
    response = fakeInputs.pop(0)
    print(f'{prompt}{response}')
    return response


def setInputs(*inputs, callerGlobals=None):
    # Set up the caller globals correctly
    if callerGlobals is None:
        callerGlobals = inspect.stack()[1][0].f_globals
    callerGlobals['input'] = fakeInput

    # Fake the inputs
    global fakeInputs
    if (len(inputs) == 1) and (isinstance(inputs[0], list) or isinstance(inputs[0], tuple)):
        fakeInputs = list(inputs[0])
    else:
        fakeInputs = list(inputs)


# An alias for graphics exercises
setTextInputs = setInputs

# Faking randrange
_randrange = random.randrange
fakeRandranges = []


def fakeRandrange(*args):
    if len(fakeRandranges) > 0:
        expected_args_and_val = fakeRandranges.pop(0)
        expected_args = expected_args_and_val[:-1]
        val = expected_args_and_val[-1]
        if args != expected_args:
            expected_arg_str = ', '.join(map(str, expected_args))
            arg_str = ', '.join(map(str, args))

            raise Exception(f'We expected to see a call to randrange({expected_arg_str}), ' +
                            f'but instead we saw a call to randrange({arg_str})')
        return val
    return _randrange(*args)


def setRandranges(randranges, callerGlobals=None):
    # Set up the caller globals correctly
    if callerGlobals is None:
        callerGlobals = inspect.stack()[1][0].f_globals
    if 'random' in callerGlobals:
        if isinstance(callerGlobals['random'], types.ModuleType):
            callerGlobals['random'].randrange = fakeRandrange
    if 'randrange' in callerGlobals:
        callerGlobals['randrange'] = fakeRandrange

    # Fake the randranges
    global fakeRandranges
    if isinstance(randranges[0], int):
        fakeRandranges = [randranges]
    else:
        fakeRandranges = list(randranges)

    for randrange in randranges:
        if not isinstance(randrange, tuple) or not (2 <= len(randrange) <= 4):
            raise Exception('setRandranges must take in a tuple of tuples with 2 to 4 elements.')

    fakeRandranges = list(randranges)


def makeTestbot(fn):
    callerGlobals = inspect.stack()[1][0].f_globals

    def testbotFn(*args, **kwargs):
        # for now, assume args are for fn, kwargs are for testbot fn
        for key in kwargs:
            if key == 'inputs':
                inputs = kwargs[key]
                if isinstance(inputs, str):
                    # if str, assume comma-separated values
                    inputs = inputs.split(',')
                setInputs(inputs, callerGlobals=callerGlobals)
            elif key == 'randranges':
                randranges = kwargs[key]
                if isinstance(randranges[0], int):
                    randranges = (randranges, )
                setRandranges(randranges, callerGlobals=callerGlobals)
            else:
                raise Exception(f'Unknown testbot keyword argument: {key}')
        startOutputCapture()
        fn(*args)
        return stopOutputCapture()

    return testbotFn


def multilineRepr(s):
    if not isinstance(s, str) or '\n' not in s:
        return repr(s)

    quote = "'"
    if "'" in s and '"' not in s:
        quote = '"'

    result = [quote * 3]

    startIdx = 0
    if s[0] == '\n':
        startIdx = 1
        result.append('\n')
    else:
        result.append('\\')
        result.append('\n')

    for i in range(startIdx, len(s)):
        c = s[i]

        if c == quote or c == '\\':
            result.append('\\')
            result.append(c)
        elif c == '\t':
            result.append('\\')
            result.append('t')
        elif c == '\n':
            result.append('\n')
        elif c == '\r':
            result.append('\\')
            result.append('t')
        elif c < ' ' or c >= '\x7f':
            result.append('\\x')
            result.append('%02x' % ord(c))
        else:
            result.append(c)

    result.append(quote * 3)

    return ''.join(result)


def getColWidths(a):
    colWidths = dict()
    for row in range(len(a)):
        if not (isinstance(a[row], list) or isinstance(a[row], tuple)):
            continue
        for col in range(len(a[row])):
            colWidths[col] = max(colWidths.get(col, 0), len(repr(a[row][col])))
    return colWidths


def nestedListReprAddElem(output, a, row, col, colWidths):
    elem = a[row][col]
    if (col > 0):
        output.append(', ')
    justMethod = 'ljust'
    if isinstance(a[row][col], int) or isinstance(a[row][col], float):
        justMethod = 'rjust'
    output.append(getattr(repr(elem), justMethod)(colWidths[col]))


def is2dList(a):
    if not isinstance(a, list):
        return False
    for elem in a:
        if isinstance(elem, list) or isinstance(elem, tuple):
            return True
    return False


def prettyListRepr(a):
    if (a == []):
        return '[]'
    if not is2dList(a):
        return repr(a)
    output = []
    colWidths = getColWidths(a)
    output.append('[\n')
    for row in range(len(a)):
        if isinstance(a[row], list) or isinstance(a[row], tuple):
            lParen, rParen = '[', ']'
            if isinstance(a[row], tuple):
                lParen, rParen = '(', ')'
            output.append(f' {lParen} ')
            for col in range(len(a[row])):
                nestedListReprAddElem(output, a, row, col, colWidths)
            output.append(f' {rParen}')
        else:
            output.append(' ' + repr(a[row]))
        if (row < len(a) - 1):
            output.append(',')
        output.append('\n')
    output.append(']')
    return ''.join(output)


def prettyStr(o):
    if isinstance(o, str) and '\n' in o:
        return multilineRepr(o)

    if isinstance(o, list):
        return prettyListRepr(o)

    return repr(o)


def prettyPrint(o):
    print(prettyStr(o))


__all__ = [
    'almostEqual', 'testFunction', 'setInputs', 'setTextInputs', 'setRandranges',
    'startOutputCapture', 'stopOutputCapture', 'prettyPrint', 'prettyStr',
    'rounded'
]
```
